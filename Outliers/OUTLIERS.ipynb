```python
import numpy as np
import pandas as pd
from IPython import display
```

```python
df=pd.read_csv('df2.csv', parse_dates=['Date'], index_col='Date')
print(len(df))
df=df['2018']
print(len(df))
df.iloc[4,4]=2000
df.iloc[3,1]=900
```

```python
def get_outliers_from_df(df): 
    outlier_indexes=[]
    for col_name in df.columns:
        q1 = df[col_name].quantile(0.25)
        q3 = df[col_name].quantile(0.75)
        iqr = q3-q1 #Interquartile range
        fence_low  = q1-1.5*iqr
        fence_high = q3+1.5*iqr
        normal_value_indexes= (df[col_name] > fence_low) & (df[col_name] < fence_high)
        liste=list(df[col_name][~normal_value_indexes].values)
        for i in liste:
            if pd.isnull(i):
                liste.remove(i)
        if liste==[]: continue
        outlier_indexes+=(liste)
    return outlier_indexes

    
def colorize(val):
    global outlier_list
    if pd.isnull(val):
        bg = '#df083a'
        color='#fff'
    elif val in outlier_list:
        bg = '#3499fa'
        color='#fff'
    else:
        bg=''
        color='#000'
    return 'background: ' + bg +'; color: '+ color

outlier_list=get_outliers_from_df(df)

df.style.applymap(colorize)
```
